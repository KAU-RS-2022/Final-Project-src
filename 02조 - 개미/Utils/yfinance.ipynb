```python
######## Stock  Data  Stream ######################################
import yfinance as yf

# Chose  your  t i c k e r
ticker_list = ["CPNG"]
path ="./data"
file_name ="stock"
start_date="2021-03-12"
last_date="2022-05-02"
for interval in ["1D"]:
    stock_data = yf.download(ticker_list,start=start_date,end=last_date,interval = interval)
    stock_data.to_csv(path+file_name + "_" + start_date + "_" + last_date  + "_"+ interval  + ".csv")

```

```python
mmm_data = pdr.get_data_yahoo(['MMM'], start='2010-01-04', end='2022-04-30')
hon_data = pdr.get_data_yahoo(['HON'], start='2022-01-04', end='2022-04-30')
kmb_data = pdr.get_data_yahoo(['KMB'], start='2022-01-01', end='2022-04-30')
apt_data = pdr.get_data_yahoo(['APT'], start='2022-01-01', end='2022-04-30')
```

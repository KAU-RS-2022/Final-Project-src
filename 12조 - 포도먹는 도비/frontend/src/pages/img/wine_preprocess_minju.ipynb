```python
import pandas as pd
import matplotlib.pyplot as plt
import os

dir = './cleansingWine.csv/'
frame = pd.read_csv('cleansingWine.csv')
pd.set_option('display.max_columns',None)

plt.rcParams["figure.figsize"] = (64,32)
fig, ax = plt.subplots(6,5)


i=0; j=0
print(frame.columns)
for c in frame.columns:
    if c in ['Unnamed: 0','id','name','producer']: continue
    if c in ['sweet', 'acidity', 'body', 'tannin']: 
        pass
    if j == 5: j = 0; i+=1
    ax[i][j].set_title(c)
    ax[i][j].bar(frame[c].value_counts().index, frame[c].value_counts().values)
    j+=1

print('start drawing...')
plt.show()


#frame.describe()
#frame.describe(include='all')
```

```python
frame = pd.read_csv('cleansingWine.csv')

frame = frame.replace(['SWEET1', 'SWEET2', 'SWEET3', 'SWEET4', 'SWEET5'], [1, 2, 3, 4, 5])
frame = frame.replace(['ACIDITY1', 'ACIDITY2', 'ACIDITY3', 'ACIDITY4', 'ACIDITY5'], [1, 2, 3, 4, 5])
frame = frame.replace(['TANNIN1', 'TANNIN2', 'TANNIN3', 'TANNIN4', 'TANNIN5'], [1, 2, 3, 4, 5])
frame = frame.replace(['BODY1', 'BODY2', 'BODY3', 'BODY4', 'BODY5'], [1, 2, 3, 4, 5])
frame = frame.replace(['Red', 'White', 'Sparkling', 'Rose'], [1, 2, 3, 4])
frame = frame.replace(['Fortified', 'Hard', 'Etc', 'Korean', 'Sake/Rice'], 5)



print("sweet")
print(frame['sweet'].value_counts())
print("acidity")
print(frame['acidity'].value_counts())
print("tannin")
print(frame['body'].value_counts())
print("type")
print(frame['type'].value_counts())
```

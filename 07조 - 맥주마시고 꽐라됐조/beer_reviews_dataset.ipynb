```python
def findMiddle(dataframe):
    # dataframe의 중간 부분을 반환하는 함수입니다.
    # return은 dataframe의 중간 10개 부분들 입니다. (dataframe 형식)
    count = dataframe.shape[0] // 2
    mid = dataframe.iloc[count - 5 : count + 5]
    
    return mid
```

```python
df_beer = df_beer.dropna() #null값 제거
df_beer.info()
print(df_beer.isnull().sum()) #null 제거 확인
```

```python
#duplication 삭제

df_beer = df_beer.sort_values('review_overall', ascending=False)

df_beer = df_beer.drop_duplicates(subset = ['review_profilename','beer_name'], keep = 'first')

df_beer.info()
```

```python
#object type -> category로 변환


df_beer["brewery_id"] = df_beer["brewery_id"].astype('category')
df_beer["beer_beerid"] = df_beer["beer_beerid"].astype('category')
df_beer["review_time"] = df_beer["review_time"].astype('category')
df_beer.describe()

#df_beer[df_beer.select_dtypes(['object']).columns] = df_beer.select_dtypes(['object']).apply(lambda x: x.astype('category'))

df_beer.info()
```

```python
#평점 0제외한 df_beer_nonspar
df_beer_nonspar = df_beer[(df_beer['review_overall'] >=1)|(df_beer['review_appearance'] >=1)]

df_beer_nonspar.info()
```

```python
df_beer_nonspar.head()
```

```python
beer_review = df_beer_nonspar[["beer_beerid","beer_name", "review_profilename","beer_abv", "review_aroma", "review_appearance", "review_palate", "review_taste", "review_overall"]]
print(beer_review.shape)

beer_review = beer_review.sort_values("review_overall", ascending=False)
beer_review.reset_index(drop=True, inplace=True)
beer_review.head(10)
```

```python
beerName = beer_review['beer_name'].unique()
print(beerName)

#맥주 종류 확인
import collections, numpy

collections.Counter(beerName)
```

```python
col = ['beer_beerid','review_profilename','review_overall']
new_df = beer_review[col]
new_df.head(500)
```

```python
#각 열 이름 변경
review = new_df.rename(index=str, columns = {'beer_beerid' : 'beer', 'review_profilename' : 'user', 'review_overall' : 'rating'})
review.head(500)
```

```python
reviewsperbeer=review.groupby('beer')['rating'].count().reset_index().sort_values(by=['rating'],ascending=False)
reviewsperbeer.head()
```

```python
# Slightly alter the figure size to make it more horizontal.
plt.figure(figsize=(8, 4), dpi=100, facecolor='w', edgecolor='k', frameon='True')
sns.set_style("darkgrid") # white/whitegrid/dark/darkgrid/ticks
sns.set_context("paper") # talk/poster

plt.hist(reviewsperbeer['rating'],alpha=0.5,color='red', bins=30)
# plot x axis label
plt.xlabel('No of Reviews per Beer', color='r')
# plot y axis label
plt.ylabel('Frequency', color='r')
# display the plot
plt.show()
```

```python
print(reviewsperbeer.loc[reviewsperbeer['rating']==1].shape[0])
print(round(100*(reviewsperbeer.loc[reviewsperbeer['rating']==1].shape[0]/reviewsperbeer.shape[0]),2))
```

```python
#리뷰 숫자에 따른 맥주 개수
#total beer Reviews : n개 -> 리뷰 수가 n개

reviewsfrequency=reviewsperbeer.groupby('rating')['beer'].count().reset_index()
reviewsfrequency.columns.values[:]=['Total Beer Reviews','Review Frequency']
reviewsfrequency.head(10)
```

```python
#review 수 50개 이상만

above50reviewsperbeer = reviewsperbeer.loc[reviewsperbeer['rating'] >= 50]
above50reviewsperbeer.columns.values[1]='Total Review Count of Beer'
above50reviewsperbeer.head()
```

```python
# Slightly alter the figure size to make it more horizontal.
plt.figure(figsize=(8, 4), dpi=100, facecolor='w', edgecolor='k', frameon='True')
sns.set_style("darkgrid") # white/whitegrid/dark/darkgrid/ticks
sns.set_context("paper") # talk/poster

plt.hist(above50reviewsperbeer['rating'],alpha=0.5,color='red', bins=30)
# plot x axis label
plt.xlabel('No of Reviews per Beer', color='r')
# plot y axis label
plt.ylabel('Frequency', color='r')
# display the plot
plt.show()
```

```python
#user 기준

reviewsperreviewer=review.groupby('user')['rating'].count().reset_index().sort_values(by=['rating'],ascending=False)
reviewsperreviewer.head()
```

```python
round(reviewsperreviewer.describe(),2).T
```

```python
#유저당 리뷰 개수 30개 이상

above30reviewsperreviewer = reviewsperreviewer.loc[reviewsperreviewer['rating'] >= 30]
above30reviewsperreviewer.columns.values[1]='Total Reviews by User'
above30reviewsperreviewer.head()
```

```python
#그래프 유저당 리뷰 개수

plt.figure(figsize=(8, 4), dpi=100, facecolor='w', edgecolor='k', frameon='True')
sns.set_style("darkgrid") # white/whitegrid/dark/darkgrid/ticks
sns.set_context("paper") # talk/poster

plt.hist(above30reviewsperreviewer['rating'],alpha=0.5,color='red', bins=30)
# plot x axis label
plt.xlabel('No of Reviews per User', color='r')
# plot y axis label
plt.ylabel('Frequency', color='r')
# display the plot
plt.show()
```

```python
beers_for_recommendation = pd.merge(review, above50reviewsperbeer, how='inner', on='beer')
beers_for_recommendation = pd.merge(beers_for_recommendation, above30reviewsperreviewer, how='inner', on='user')
beers_for_recommendation.shape
```

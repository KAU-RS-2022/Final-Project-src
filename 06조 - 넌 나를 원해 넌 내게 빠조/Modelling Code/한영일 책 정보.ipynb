```python
import json
import pandas as pd
import numpy as np
from collections import Counter
```

# json 파일 로드 후 공백 제거 등 진행

```python
with open('한국소설_책_정보.json','r', encoding="UTF-8") as f:
    kr_data = json.load(f)
for book in kr_data:
    book['country'] = '한국'
    
with open('영미소설_책_정보.json','r', encoding="UTF-8") as f:
    uk_data = json.load(f)
for book in uk_data:
    book['country'] = '영미'
    
with open('일본소설_책_정보.json','r', encoding="UTF-8") as f:
    jp_data = json.load(f)
for book in jp_data:
    book['country'] = '일본'
    
info = kr_data + uk_data + jp_data
```

```python
for book in info:
    book["title"] = book["title"].split('(')[0].strip()
    book["author"] = book["author"].split(' ')[0].split(',')[0]
    book["publisher"] = book["publisher"].strip()
```

근데 위의 title, book_id 쌍 리스트에서 'title'이 동일한 책들이 존재하며, 또한 한국소설, 영미소설, 일본소설에 동일제목+동일ID의 책이 동시에 존재한다. 즉, 총 데이터에서 동일한 제목과 동일한 book_id 값이 존재하게 되는것이다. 따라서, 본 팀은 동일한 'title'일 시 가장 먼저 나오는 'title' 에 대해 book_id를 통일할 것이며, 동일한 'book_id'일 시에도 가장 먼저 나오는 book_id(title)에 대해 값을 통일 시켜 줄 것이다.

### 동일 title 몇개, 동일 book_id 몇개

```python
book_id_list = []
for book in info:
    book_id_list.append({"title": book['title'], "book_id" : book["book_id"]}) 
print(len(book_id_list))
```

```python
title_list = []
for book in book_id_list:
    title_list.append(book['title'])
dup_title_list = [] # 동일한 책 제목인 책들
result = Counter(title_list)
for key, value in result.items():
    if value >= 2: # 2개이상 책 제목을 가지고 있는 책 들
        dup_title_list.append(key)
print(len(dup_title_list)) # 3017개 책들이 2개이상의 동일한 책 제목을 가짐
dup_title_list
```

```python
id_list=[]
for  book in book_id_list:
    id_list.append(book['book_id'])
dup_id_list = [] # 동일한 id인 책들
result = Counter(id_list)
for key, value in result.items():
    if value >= 2:
        dup_id_list.append(key)
print(len(dup_id_list)) # 43개의 책이 2개 이상의 동일한 책 id를 가짐
dup_id_list
```
